```python
import numpy as np

data1 = np.array([1, 3, 5, 7, 9])
data2 = np.array([3, 4, 5, 6, 7])
```

```python
data1.mean(), data2.mean()
```

```python
def get_var(data):
    data_mean = data.mean()
    
    total = 0
    
    for d in data:
        total += (d - data_mean)**2
        
    return total/len(data)

# 분산 공식을 코드화 한 것
    
d1_var = get_var(data1)
d2_var = get_var(data2)
```

```python
d1_var, d2_var
```

```python
d1_std = np.sqrt(d1_var)
d2_std = np.sqrt(d2_var)

d1_std, d2_std
```

```python
data1.var(), data2.var()
```
