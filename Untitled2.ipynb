```python
import pandas as pd
```

```python
df = pd.read_csv("./data/korean-idol.csv")
df.head()
```

apply
Series, DataFrame에 구체적이고 복잡한 로직에 의해 데이터를 편집하고 싶을 때 사용
    단순 집계 뿐만이 아닌 각 데이터에 프로그래밍적인 효과
        데이터에 대한 여러 조건 검사
        형태소 분리
        복잡한 조건에 의한 데이터 편집 등
    apply(함수)
        apply 안에 들어가는 함수는 데이터 편집 후 반드시 리턴

```python
df2 = df.copy()

man_mask = df2["성별"] == "남자"
woman_mask = df2["성별"] == "여자"

df2.loc[man_mask, "성별"] = 1
df2.loc[woman_mask, "성별"] = 0

df2.head()
```

```python
df["성별_new"] = df["성별"].apply(lambda x : 1 if x == "남자" else 0)
df.head()
```

```python
df["키/2"] = df["키"].apply(lambda x : x / 2)
df.head()
```
