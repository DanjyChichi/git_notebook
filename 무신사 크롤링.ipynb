```python
for item in best_list:
    item_name = item.select_one("p.list_info > a").text
    params = item_name.strip()
    item_name = params
    item_brand = item.select_one("p.item_title > a").text
    item_link = item.select_one("p.list_info > a").get("href")
    
    

    parsed_url = urlparse(item_link)
    url_query = parsed_url.path.split("/")
    url_query = url_query[2 : 4]
    url_query = "=".join(url_query)
    parsed_query = parse_qs(url_query)
    item_code = parsed_query["goods"][0]
    
    img_url = item.select_one("#goodsRankList > li.li_box > div.li_inner > div.list_img > a > img").get("data-original")
    urlretrieve(img_url, "./musinsa_images/"+item_code+".jpg")
    
    if item.select_one("p.point > span.img-score > span") == None:
        pass
    else:
        item_grade = item.select_one("p.point > span.img-score > span").get("style")
        item_grade = item_grade.split(" ")
        item_grade = item_grade[1].replace("%", "")
        item_grade = float(item_grade) / 20
        item_grade
    
    print("상품명 :", item_name, "브랜드 :", item_brand, "\n", "제품 링크 :", item_link, "별점", item_grade,"점", "상품 코드 :", item_code)
```
